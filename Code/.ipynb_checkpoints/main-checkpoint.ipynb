```python
import pandas as pd
import numpy as np
```

```python
#Loading train data into X_train and y_train
df_train = pd.read_csv("../Dataset/fashion-mnist_train.csv")
columns_train = df_train.columns
X_train = df_train.filter(items = columns_train[1:])
y_train = df_train.filter(items = columns_train[:1])

#print(X_train.head)
#print(y_train.head)
```

```python
#now it's time to implement PCA
def Manual_PCA(X, n):
    #Creating n*d matrix
    tmpX = np.array(X)
    #Subtracting mean
    tmpX = tmpX - np.mean(tmpX, axis = 0)
    #Calculating covariance matrix
    tmpX_cov = np.cov(tmpX.T)
    #Finding eigen values and eigen vectors
    tmpX_eig_val, tmpX_eig_vec = np.linalg.eig(tmpX_cov)
    #Sorting eigen vectors base on eigen values
    for i in range(1, tmpX_eig_vec.shape[0]):
        cur = i
        flag = tmpX_eig_val[cur] > tmpX_eig_val[cur - 1]
        while(cur >= 1 and flag):
            tmpX_eig_vec[:, [cur, cur - 1]] = tmpX_eig_vec[:, [cur - 1, cur]]
            tmpX_eig_val[[cur, cur - 1]] = tmpX_eig_val[[cur - 1, cur]]
            cur -= 1
            if(cur >= 1):
                flag = tmpX_eig_val[cur] > tmpX_eig_val[cur - 1]
    #Selecting first n eigen vectors
    tmpX_sel_eig_vec = tmpX_eig_vec[:,:n]
    return np.dot(tmpX, tmpX_sel_eig_vec), tmpX_sel_eig_vec
```

```python
manually_transformed_3d_X_train, manually_fitted_3d_pca = Manual_PCA(X_train, 3)
manually_transformed_3d_X_test = np.dot(X_test - np.mean(X_train, axis = 0), manually_fitted_3d_pca)

manually_transformed_2d_X_train, manually_fitted_2d_pca = Manual_PCA(X_train, 2)
manually_transformed_2d_X_test = np.dot(X_test - np.mean(X_train, axis = 0), manually_fitted_2d_pca)
```

```python
#Now lets implement pca with scikit-learn library
from sklearn.decomposition import PCA
def library_PCA(X, n):
    #Creating n*d matrix
    tmpX = X
    #Subtracting mean
    tmpX = tmpX - np.mean(tmpX, axis = 0)
    #Builiding pca
    pca = PCA(n_components=n)
    #Finding Components
    principalComponents = pca.fit_transform(tmpX)
    return principalComponents, pca
    
```

```python
lib_transformed_3d_X_train, lib_fitted_3d_pca = library_PCA(X_train, 3)
lib_transformed_3d_X_test = lib_fitted_3d_pca.transform(X_test)

lib_transformed_2d_X_train, lib_fitted_2d_pca = library_PCA(X_train, 2)
lib_transformed_2d_X_test = lib_fitted_2d_pca.transform(X_test)
```

```python
#Now its time to learn SVM
from sklearn import svm
from sklearn import metrics
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import cross_validate
from sklearn.utils import shuffle

def EQ_Filter(X, y, sz):
    ansx = []
    ansy = []
    cur = 0
    cnt = 0
    cntc = 10 * [0]
    while(cnt < sz):
        if(cntc[y.iloc[cur][0]] < (sz // 10)):
            ansy.append(y.iloc[cur][0])
            ansx.append(np.array(X.iloc[cur]))
            cntc[y.iloc[cur][0]] += 1
            cnt += 1
        cur += 1
    return np.array(ansx), np.array(ansy)
        

SVM_train = X_train
SVM_train["lable"] = y_train
SVM_train = shuffle(SVM_train)
SVM_train_columns = SVM_train.columns
SVM_X_train = SVM_train.filter(items = SVM_train_columns[0:-1])
SVM_y_train = SVM_train.filter(items = SVM_train_columns[-1:])

PCA_components = [3, 5, 10, 20, 40]
Sample_Sizes = [5000, 10000, 20000]
Kernels = ['linear', 'poly', 'rbf', 'sigmoid']
Degrees = [3, 4]

best_PCA_component = 0
best_Kernel = ''
best_Sample_Size = 0
best_Degree = 0
best_score = 0

for Sample_Size in Sample_Sizes:
    for PCA_component in PCA_components:
        SVM_samples_X, SVM_samples_Y = EQ_Filter(SVM_X_train, SVM_y_train, Sample_Size)
        lib_transformed_custom_X_train, lib_fitted_custom_pca = library_PCA(SVM_samples_X, PCA_component)
        lib_transformed_custom_X_test = lib_fitted_custom_pca.transform(X_test)
        for Kernel in Kernels:
            if(Kernel == 'poly'):
                for Degree in Degrees:
                    my_pipeline = Pipeline(steps=[('preprocessor', SimpleImputer()),
                                                  ('model', svm.SVC(kernel = Kernel,
                                                            degree = Degree))
                                                 ])
                    print("#PCA_component: ", PCA_component,
                         "| Kernel: ", Kernel,
                         "| Degree:", Degree,
                         "| Sample_Size: ", Sample_Size)
                    scores = cross_validate(my_pipeline, lib_transformed_custom_X_train, 
                                    SVM_samples_Y,
                                    cv=5,
                                    scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'],
                                    return_train_score=True, return_estimator=True)
                    accurancy_scores = scores['test_accuracy']
                    precision_scores = scores['test_precision_macro']
                    recall_scores = scores['test_recall_macro']
                    f1_scores = scores['test_f1_macro']
                    print("accurancy_scores:\n", accurancy_scores)
                    print("precision_scores:\n", precision_scores)
                    print("recall_scores:\n", recall_scores)
                    print("f1_scores:\n", f1_scores)
                    score = np.mean(accurancy_scores)
                    print("Average accurancy score:", score)
                    if(score > best_score):
                        best_PCA_component = PCA_component
                        best_Kernel = Kernel
                        best_Sample_Size = Sample_Size
                        best_Degree = Degree
                        best_score = score

            else:
                my_pipeline = Pipeline(steps=[('preprocessor', SimpleImputer()),
                                              ('model', svm.SVC(kernel = Kernel))
                                             ])
                print("#PCA_component: ", PCA_component,
                         "| Kernel: ", Kernel,
                         "| Sample_Size:", Sample_Size)
                scores = cross_validate(my_pipeline, lib_transformed_custom_X_train, 
                                        SVM_samples_Y,
                                        cv=5,
                                        scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'],
                                        return_train_score=True, return_estimator=True)
                accurancy_scores = scores['test_accuracy']
                precision_scores = scores['test_precision_macro']
                recall_scores = scores['test_recall_macro']
                f1_scores = scores['test_f1_macro']
                print("accurancy_scores:\n", accurancy_scores)
                print("precision_scores:\n", precision_scores)
                print("recall_scores:\n", recall_scores)
                print("f1_scores:\n", f1_scores)
                score = np.mean(accurancy_scores)
                print("Average accurancy score:", score)
                if(score > best_score):
                    best_PCA_component = PCA_component
                    best_Kernel = Kernel
                    best_Sample_Size = Sample_Size
                    best_score = score

print("best_score", best_score)
print("best_Sample_Size", best_Sample_Size)
print("best_Kernel", best_Kernel)
print("best_PCA_component", best_PCA_component)
print("best_Degree", best_Degree)
```

```python
from sklearn.naive_bayes import GaussianNB

X_train = df_train.filter(items = columns_train[1:])

PCA_components = [3, 5, 10, 20, 40, 80, 160, 320]
best_PCA_component = 0
best_score = 0

for PCA_component in PCA_components:
    lib_transformed_custom_X_train, lib_fitted_custom_pca = library_PCA(X_train, PCA_component)
    lib_transformed_custom_X_test = lib_fitted_custom_pca.transform(X_test)
    my_pipeline = Pipeline(steps=[('preprocessor', SimpleImputer()),
                          ('model', GaussianNB())
                         ])
    scores = cross_validate(my_pipeline, lib_transformed_custom_X_train, 
            np.array([x[0] for x in np.array(y_train)]),
            cv=5,
            scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'],
            return_train_score=True, return_estimator=True)
    accurancy_scores = scores['test_accuracy']
    precision_scores = scores['test_precision_macro']
    recall_scores = scores['test_recall_macro']
    f1_scores = scores['test_f1_macro']
    print("#PCA_component: ", PCA_component)
    print("accurancy_scores:\n", accurancy_scores)
    print("precision_scores:\n", precision_scores)
    print("recall_scores:\n", recall_scores)
    print("f1_scores:\n", f1_scores)
    score = np.mean(accurancy_scores)
    print("Average accurancy score:", score)
    if(score > best_score):
        best_PCA_component = PCA_component
        best_score = score
print("best #pca component : ", best_PCA_component)
print("best score: ", best_score)
    
```

```python
from sklearn.neighbors import KNeighborsClassifier

PCA_components = [3, 5, 10, 20, 40, 80, 160, 320]
N_neighbors = [3, 5, 7]
Weights = ['uniform', 'distance']
best_PCA_component = 0
best_N_neighbor = 0
best_Weight =''
best_score = 0

for PCA_component in PCA_components:
    lib_transformed_custom_X_train, lib_fitted_custom_pca = library_PCA(X_train, PCA_component)
    lib_transformed_custom_X_test = lib_fitted_custom_pca.transform(X_test)
    for N_neighbor in N_neighbors:
        for Weight in Weights:
            my_pipeline = Pipeline(steps=[('preprocessor', SimpleImputer()),
                                  ('model', KNeighborsClassifier(n_neighbors=N_neighbor,
                                                                weights=Weight))
                                 ])
            scores = cross_validate(my_pipeline, lib_transformed_custom_X_train, 
                    np.array([x[0] for x in np.array(y_train)]),
                    cv=5,
                    scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'],
                    return_train_score=True, return_estimator=True)
            accurancy_scores = scores['test_accuracy']
            precision_scores = scores['test_precision_macro']
            recall_scores = scores['test_recall_macro']
            f1_scores = scores['test_f1_macro']
            print("#PCA_component: ", PCA_component)
            print("N_neghbor: ", N_neighbor)
            print("Weight: ", Weight)
            print("accurancy_scores:\n", accurancy_scores)
            print("precision_scores:\n", precision_scores)
            print("recall_scores:\n", recall_scores)
            print("f1_scores:\n", f1_scores)
            score = np.mean(accurancy_scores)
            print("Average accurancy score:", score)
            if(score > best_score):
                best_PCA_component = PCA_component
                best_N_neighbor = N_neighbor
                best_Weight = Weight
                best_score = score
print("best #pca component : ", best_PCA_component)
print("best score: ", best_score)
```

```python
from sklearn.neural_network import MLPClassifier

PCA_components = [10]
Hidden_layer_sizes = [(100,), (100, 25)]
Learning_rates = ['constant', 'invscaling', 'adaptive']
Activations = ['identity', 'logistic', 'tanh', 'relu']
#Batch_sizes = [200, 300]
#Max_iters = [200, 250]

best_PCA_components = 0
best_Hidden_layer_size = 0
best_Learning_rate = 0
best_Activation = 0
#best_Batch_size = 0
#best_max_iter = 0
best_score = 0

for PCA_component in PCA_components:
    lib_transformed_custom_X_train, lib_fitted_custom_pca = library_PCA(X_train, PCA_component)
    lib_transformed_custom_X_test = lib_fitted_custom_pca.transform(X_test)
    for Hidden_layer_size in Hidden_layer_sizes:
        for Learning_rate in Learning_rates:
            for Activation in Activations:
                #for Batch_size in Batch_sizes:
                    #for Max_iter in Max_iters:
                my_pipeline = Pipeline(steps=[('preprocessor', SimpleImputer()),
                                      ('model', MLPClassifier(hidden_layer_sizes=Hidden_layer_size,
                                                             learning_rate=Learning_rate,
                                                             activation = Activation))
                                                             #batch_size = Batch_size,
                                                             #max_iter = Max_iter))
                                     ])
                scores = cross_validate(my_pipeline, lib_transformed_custom_X_train, 
                        np.array([x[0] for x in np.array(y_train)]),
                        cv=5,
                        scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'],
                        return_train_score=True, return_estimator=True)
                accurancy_scores = scores['test_accuracy']
                precision_scores = scores['test_precision_macro']
                recall_scores = scores['test_recall_macro']
                f1_scores = scores['test_f1_macro']
                print("#PCA_component: ", PCA_component)
                print("Hidden layer size: ", Hidden_layer_size)
                print("Learning rate: ", Learning_rate)
                print("Activation: ", Activation)
                #print("Batch size: ", Batch_size)
                #print("Max iter: ", Max_iter)
                print("accurancy_scores:\n", accurancy_scores)
                print("precision_scores:\n", precision_scores)
                print("recall_scores:\n", recall_scores)
                print("f1_scores:\n", f1_scores)
                score = np.mean(accurancy_scores)
                print("Average accurancy score:", score)
                if(score > best_score):
                    best_PCA_components = PCA_components
                    best_Hidden_layer_size = Hidden_layer_size
                    best_Learning_rate = Learning_rate
                    best_Activation = Activation
                    #best_Batch_size = Batch_size
                    #best_max_iter = Max_iter
                    best_score = score
print("best #pca component : ", best_PCA_component)
print("best Hidden layer size: ", best_Hidden_layer_size)
print("best Learning rate: ", best_Learning_rate)
print("best Activation: ", best_Activation)
#print("best Batch size: ", best_Batch_size)
#print("best Max iter: ", best_Max_iter)
print("best score: ", best_score)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix


#SVM
SVM_samples_X, SVM_samples_Y = EQ_Filter(SVM_X_train, SVM_y_train, 20000)
lib_transformed_custom_X_train, lib_fitted_custom_pca = library_PCA(SVM_samples_X, 40)
lib_transformed_custom_X_test = lib_fitted_custom_pca.transform(X_test - np.mean(X_test, axis = 0))

model = svm.SVC(kernel = 'rbf')
model.fit(lib_transformed_custom_X_train, SVM_samples_Y)
y_pred = model.predict(lib_transformed_custom_X_test)
y_pred_train = model.predict(lib_transformed_custom_X_train)

accuracy = accuracy_score(y_test, y_pred)
accuracy_train = accuracy_score(SVM_samples_Y, y_pred_train)

precision = precision_score(y_test, y_pred, average = "macro")
precision_train = precision_score(SVM_samples_Y, y_pred_train, average = "macro")

recall = recall_score(y_test, y_pred, average = "macro")
recall_train = recall_score(SVM_samples_Y, y_pred_train, average = "macro")

f1 = f1_score(y_test, y_pred, average = "macro")
f1_train = f1_score(SVM_samples_Y, y_pred_train, average = "macro")

confusion = confusion_matrix(y_test, y_pred)
confusion_train = confusion_matrix(SVM_samples_Y, y_pred_train)

print("accuracy", accuracy)
print("accuracy_train", accuracy_train)
print("precision",precision)
print("precision_train",precision_train)
print("recall", recall)
print("recall_train", recall_train)
print("f1", f1)
print("f1_train", f1_train)
print("confusion", confusion)
print("confision_train", confusion_train)
```
